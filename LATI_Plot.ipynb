```python
from scipy.stats import qmc
from math import *
import numpy as np
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt

a0 = 0.1
b0 = 1
a1 = 2
b1 = 1
sample_points_2_index = 8 #采样点，2^n个point

sampler = qmc.LatinHypercube(d=3)
#sampler = qmc.LatinHypercube(d=3,optimization="random-cd")
sample = sampler.random(n=2**sample_points_2_index)

l_bounds = -3.14
u_bounds = 3.14
fianl_sample = qmc.scale(sample, l_bounds, u_bounds)
fianl_sample.shape
```

```python
fig = plt.figure(figsize=(10,10),facecolor='white')
ax = fig.add_subplot(111, projection='3d')


ax.scatter(fianl_sample[:,0], fianl_sample[:,1],fianl_sample[:,2])

ax.set_xlabel('X Label')
ax.set_ylabel('Y Label')
ax.set_zlabel('Z Label')


plt.savefig("lati.svg", dpi=300,format="svg")  # 注意这个必须放在show之前！！不然svg只有2kb，啥也没有
plt.show()
```
