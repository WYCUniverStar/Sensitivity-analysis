```python
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits.mplot3d import Axes3D #3D
 
x = np.linspace(-1,1,100)
y = np.linspace(-1,1,100)
x_, y_ = np.meshgrid(x,y,indexing='ij')
z_ = x_**2 + y_**2 + x_ / 3 + y_ / 3  # 
fig = plt.figure(figsize=(10, 10), facecolor='white') #
# sub = fig.add_subplot(111, projection='3d')# 
sub = fig.add_subplot( projection='3d')# 
surf = sub.plot_surface(x_, y_, z_, cmap=plt.cm.brg) #,cmap=plt.cm.brgcmap
cb = fig.colorbar(surf, shrink=0.8, aspect=15) #
 
sub.set_xlabel(r"x axis")
sub.set_ylabel(r"y axis")
sub.set_zlabel(r"z axis")

plt.savefig("test.svg", dpi=300,format="svg")
plt.show()


z_.shape
```
